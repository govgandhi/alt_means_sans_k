```python
from pyclustering.cluster import cluster_visualizer
from pyclustering.cluster.xmeans import xmeans, splitting_type
from pyclustering.cluster.center_initializer import kmeans_plusplus_initializer
from pyclustering.utils import read_sample
from pyclustering.samples.definitions import FCPS_SAMPLES
import numpy as np
import pyclustering as pct
print(np. __version__ )
print(pct. __version__ )
```

```python
# Read sample 'Target'.
sample = read_sample(FCPS_SAMPLES.SAMPLE_TARGET)
 
# Prepare initial centers - amount of initial centers defines amount of clusters from which X-Means will start analysis.
random_seed = 1000
amount_initial_centers = 3
initial_centers = kmeans_plusplus_initializer(sample, amount_initial_centers, random_state=random_seed).initialize()
 
# Create instance of X-Means algorithm with MNDL splitting criterion.
xmeans_mndl = xmeans(sample, initial_centers, 20, splitting_type=splitting_type.MINIMUM_NOISELESS_DESCRIPTION_LENGTH, random_state=random_seed)
xmeans_mndl.process()
 
# Extract X-Means MNDL clustering results.
mndl_clusters = xmeans_mndl.get_clusters()
 
# Visualize clustering results.
visualizer = cluster_visualizer(titles=['X-Means with MNDL criterion'])
visualizer.append_clusters(mndl_clusters, sample)
visualizer.show()

"Fin"
```

Okay, we have the different embeddings, a comparison of values for which performs the best for our proposed method. Let's put it into a function that returns the labels

```python
def get_proposed_labels(emb):
    rpos, cpos, vpos = find_knn_edges(emb, num_neighbors=100, device = "cuda:3")
    cneg = np.random.choice(emb.shape[0], len(cpos))
    vneg = np.array(np.sum(emb[rpos, :] * emb[cneg, :], axis=1)).reshape(-1)

    model = LogisticRegression()
    model.fit(
        np.concatenate([vpos, vneg]).reshape((-1, 1)),
        np.concatenate([np.ones_like(vpos), np.zeros_like(vneg)]),
            )
    w1, b0 = model.coef_[0, 0], -model.intercept_[0] 
    proposed_labels = louvain(emb, w1, b0, device = "cuda:3")

    return proposed_labels

for method in methods:
    print(f"Method {method} has esim score: ", (calc_esim(community_table["community_id"], get_proposed_labels(embeddings[method]))))
    
```
